#### 2.2. Masking

```python
apple.head()
```

```python
apple[apple.Close < 0.848214]
```

```python
apple.query('Close < 0.848214')
```

```python
apple[(apple.Close < 0.848214) & (apple.Volume < 740667200)]
```

```python
# apple[(apple.Close < 0.848214) & (apple.Volume < 740667200)]
apple.query('Close < 0.848214 & Volume < 740667200')
```

```python
apple.query('Close < 0.848214 & Volume < 740667200')[['Open', 'Close']]
```

```python
apple.sort_values(by=['Open'], ascending=False)
```
