```python

import pandas as pd
import numpy as np
```

```python

import matplotlib.pyplot as plt
```

```python
import seaborn as sns
from seaborn import load_dataset
```

```python
tips = load_dataset('tips')
iris = load_dataset('iris')
titanic = load_dataset('titanic')
```

```python
tips.head(2)
```

```python
sns.scatterplot(data=iris, x='sepal_length', y= 'sepal_width', hue ='species');
```

```python
setosa = iris.query('species== "setosa"')
```

```python
sns.scatterplot(data= setosa,x='sepal_length', y= 'sepal_width');
```

```python
sns.displot(data = tips,x='total_bill');
```

```python
fig,ax = plt.subplots (1,2, figsize= (15,5))
sns.histplot(data = tips,x= 'total_bill', ax = ax[0])
sns.scatterplot (data =tips, x= 'total_bill', ax = ax[1], hue = 'sex');
```

```python
g=sns.FacetGrid( data = tips, row = 'sex', col = 'smoker',hue='time')
g.map(sns.scatterplot, 'total_bill','tip')
plt.legend();
```

```python
sns.catplot ( data = tips, x='sex', y='total_bill',row = 'smoker',col = 'time', kind = 'strip');
```

```python
sns.catplot ( data = tips, x='sex',row = 'smoker',col = 'time', kind = 'count');
```

```python
sns.catplot ( data = tips, x='sex', y='total_bill',row = 'smoker',col = 'time', kind = 'box');
```

```python
sns.lmplot (data = tips, x= 'total_bill', y='tip');
```
