```python
sql = "SELECT e.EMP_ID, concat(e.LAST_NAME, ' ' ,e.FIRST_NAME) AS FULL_NAME, D.NAME AS DEPT_NAME \
FROM EMPLOYEE e INNER JOIN DEPARTMENT D on e.DEPT_ID = D.DEPT_ID \
INNER JOIN BRANCH B on e.ASSIGNED_BRANCH_ID = B.BRANCH_ID;"
employee = pd.read_sql(sql, connection)
account = pd.read_sql("SELECT * FROM ACCOUNT", connection)
```
